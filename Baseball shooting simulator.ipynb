```python
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

flight_data = []
athletes_data = []

while True:
    print("\nBu program, verilen bir baseball topunun uçuş süresini, menzilini ve maksimum yüksekliğini hesaplar")
    print("Lütfen bir seçenek seçin:")
    print("1. Uçuş Süresini Hesapla")
    print("2. Menzili Hesapla")
    print("3. Maksimum Yüksekliği Hesapla")
    print("4. 3D Grafik Göster")
    print("5. Çıkış")

    choice = input("Seçiminizi yapın (1/2/3/4/5): ")

    if choice == "1":
        #Let's get the calculate flight time.
        athlete_name = input("Sporcu adını girin: ")  
        initial_velocity = float(input("Başlangıç hızını metre/sn cinsinden giriniz: "))
        azimuth = int(input("Atış açısını derece cinsinden giriniz: "))
        
        
        # Let's add error checking
        if initial_velocity < 0 or azimuth < 0 or azimuth > 90:
            print("Hatalı giriş! Başlangıç hızı pozitif olmalı ve açı 0 ile 90 derece arasında olmalıdır.")
        else:
            G = 9.80
            initial_velocity_x = initial_velocity * math.cos(math.radians(azimuth))
            initial_velocity_y = initial_velocity * math.sin(math.radians(azimuth))
            flight_time = (2 * initial_velocity_y) / G
            
            
            # add data to the list
            flight_data.append({
                "Sporcu": athlete_name,  # Doğru sporcu adını kullanın
                "Uçuş Türü": "Uçuş Süresi",
                "Başlangıç Hızı": initial_velocity,
                "Atış Açısı": azimuth,
                "Uçuş Süresi": flight_time
            })
            
            
            # Flight duration and other details
            print(f"Uçuş Süresi: {flight_time:.3f} saniye")
            print(f"Başlangıç Hızı X Bileşeni: {initial_velocity_x:.3f} m/s")
            print(f"Başlangıç Hızı Y Bileşeni: {initial_velocity_y:.3f} m/s")
            
            
    elif choice == "2":
        athlete_name = input("Sporcu adını girin: ")  
        # Let's get calculate azimuth
        initial_velocity = float(input("Başlangıç hızını metre/sn cinsinden giriniz: "))
        azimuth = int(input("Atış açısını derece cinsinden giriniz: "))
        
        # Let's add error checking
        if initial_velocity < 0 or azimuth < 0 or azimuth > 90:
            print("Hatalı giriş! Başlangıç hızı pozitif olmalı ve açı 0 ile 90 derece arasında olmalıdır.")
        else:
            G = 9.80
            initial_velocity_x = initial_velocity * math.cos(math.radians(azimuth))
            initial_velocity_y = initial_velocity * math.sin(math.radians(azimuth))
            flight_time = (2 * initial_velocity_y) / G
            destination = initial_velocity_x * flight_time
            
            
            # add data to the list
            flight_data.append({
                "Sporcu": athlete_name,  
                "Uçuş Türü": "Menzil Hesabı",
                "Başlangıç Hızı": initial_velocity,
                "Atış Açısı": azimuth,
                "Menzil": destination,
                "Uçuş Süresi": flight_time
            })
            
                        # destination and other details
            print(f"Menzil: {destination:.3f} metre")
            print(f"Uçuş süresi: {flight_time:.3f} saniye ")
            
            
    elif choice == "3":
        athlete_name = input("Sporcu adını girin: ")  
        # Calculate maximum height
        initial_velocity = float(input("Başlangıç hızını metre/sn cinsinden giriniz: "))
        azimuth = int(input("Atış açısını derece cinsinden giriniz: "))
        
        
        # add error checking
        if initial_velocity < 0 or azimuth < 0 or azimuth > 90:
            print("Hatalı giriş! Başlangıç hızı pozitif olmalı ve açı 0 ile 90 derece arasında olmalıdır.")
        else:
            G = 9.80            
            initial_velocity_y = initial_velocity * math.sin(math.radians(azimuth))
            maximum_height = (initial_velocity_y ** 2) / (2 * G)
            
            
            # add data to the list
            flight_data.append({
                "Sporcu": athlete_name,  # Doğru sporcu adını kullanın
                "Uçuş Türü": "Maksimum Yükseklik Hesabı",
                "Başlangıç Hızı": initial_velocity,
                "Atış Açısı": azimuth,
                "Maksimum Yükseklik": maximum_height,
                "Başlangıç Hızı Y bileşeni": initial_velocity_y
            })
            
            
            # maximum height and other details
            print(f"Maksimum Yükseklik: {maximum_height:.3f} metre")
            print(f"Başlangıç Hızı Y bileşeni: {initial_velocity_y:.3f} m/s")
    
    elif choice == "4":
        # Show to the 3d graphic
        if len(flight_data) == 0:
            print("Önce uçuş verilerini hesaplayın (Seçenek 1, 2 veya 3)!")
        else:
            data = flight_data[-1]  #uses the last calculate data
            t = np.linspace(0, data["Uçuş Süresi"], num=100)
            x = initial_velocity_x * t
            y = initial_velocity_y * t - 0.5 * G * t ** 2
            z = np.zeros_like(t)

            fig = plt.figure()
            ax = fig.add_subplot(111, projection="3d")
            ax.plot(x, y, z)
            ax.set_xlabel("X (m)")
            ax.set_ylabel("Y (m)")
            ax.set_zlabel("Z (m)")
            ax.set_title("3D Top Uçuşu")
            plt.show()

    elif choice == "5":
        print("Programdan çıkılıyor. ")

        # Save data to a file
        with open("flight_data.txt", "w") as file:
            for data_point in flight_data:
                file.write(f"Sporcu: {data_point['Sporcu']}\n")
                file.write(f"Uçuş Türü: {data_point['Uçuş Türü']}\n")
                file.write(f"Başlangıç Hızı: {data_point['Başlangıç Hızı']} m/s\n")
                file.write(f"Atış Açısı: {data_point['Atış Açısı']} derece\n")
                if 'Menzil' in data_point:
                    file.write(f"Menzil: {data_point['Menzil']:.3f} metre\n")
                if 'Maksimum Yükseklik' in data_point:
                    file.write(f"Maksimum Yükseklik: {data_point['Maksimum Yükseklik']:.3f} metre\n")
                if 'Uçuş Süresi' in data_point:
                    file.write(f"{data_point['Uçuş Türü']}: {data_point['Uçuş Süresi']:.3f} saniye\n")
                file.write("\n")

        # Exit the program
        break

```
